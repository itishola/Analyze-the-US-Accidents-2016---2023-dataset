## Analyze the US Accidents 2016 - 2023 dataset

#### This is a countrywide car accident dataset that covers 49 states of the USA. The accident data were collected from February 2016 to March 2023, using multiple APIs that provide streaming traffic incident (or event) data. These APIs broadcast traffic data captured by various entities, including the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors within the road networks. The dataset currently contains approximately 7.7 million accident records. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('US_Accidents_March23.csv')
print("The dataset shape:", df.shape)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Convert Start_Time to datetime type with a more flexible approach
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='mixed', errors='coerce')

# Now create the Year column
df['Year'] = df['Start_Time'].dt.year  # Create a Year column

# Get all accident records - note that year is now numeric, not string
accident_records = df[df['Year'].between(2016, 2023)]  # Using integers instead of strings
```

```python
# Extract locations
locations = accident_records['State']
```

```python
# Count the frequency of each location
location_counts = locations.value_counts()
```

```python
# Create and display bar plot
location_counts.plot(kind='bar')
plt.title('Most Common Accident Locations in the US')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

The bar chart displayed showing that CA, NY, and SC are among the most common accident locations.

### Let's check the Accident Severity by Weather Conditions

```python
location_by_weather = accident_records.groupby(['State', 'Year', 'Weather_Condition']).size().reset_index(name='count')
```

```python
# Create a severity mapping for weather conditions
severity_map = {
    'Good': 1,
    'Fair': 2,
    'Poor': 3
}
```

```python
print("Unique weather values:", accident_records['Weather_Condition'].unique())
```

```python
# Apply mapping (assuming the Weather column contains the keys in severity_map)
location_by_weather['Severity'] = location_by_weather['Weather_Condition'].map(severity_map)
```

```python
# 4. Create a heat map to visualize the correlation
plt.figure(figsize=(10, 8))

# Create a pivot table for the heatmap
pivot_data = location_by_weather.pivot_table(
    index='State', 
    columns='Year', 
    values='Severity',  
    aggfunc='mean'      
).fillna(0)

# Create the heatmap
sns.heatmap(
    pivot_data,
    cmap='Blues',
    annot=True,
    fmt='.1f',
    cbar=True,
    vmin=0
)

plt.title('Weather Clarity by Location Correlation', fontsize=12, y=1.05)
plt.tight_layout()
plt.show()
```

##### The heat map displayed showing that both rain and snow contribute to more severe accidents.

#### Let's check the Time of Day with Most Accidents

```python
# 2. Extract hour from End_Time column
# First, ensure End_Time is in datetime format
accident_records['End_Time'] = pd.to_datetime(accident_records['End_Time'], errors='coerce')

# Extract hour from the datetime
accident_records['Hour'] = accident_records['End_Time'].dt.hour
```

```python
# Group by location and hour, then count accidents
hourly_counts = accident_records.groupby(['State', 'Hour']).size().reset_index(name='Count')
```

```python
# Plot a line chart showing the trend over different times of the day
plt.figure(figsize=(12, 15))

# Create a line for each state
sns.lineplot(x='Hour', y='Count', hue='State', data=hourly_counts)
```

```python
overall_hourly = hourly_counts.groupby('Hour')['Count'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.lineplot(x='Hour', y='Count', data=overall_hourly)

plt.title('Accident Frequency by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(0, 24), [f"{h}:00" for h in range(24)])
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### The line chart displayed showing that morning and evening hours have the highest accident frequencies.

##### Let's check the Correlation Between Visibility and Accident Severity

```python
# Group by location and year, then compute the average visibility and severity
grouped_data = accident_records.groupby(['State', 'Year']).agg({
    'Visibility(mi)': 'mean',
    'Severity': 'mean'
}).reset_index()
```

```python
# Print the first few rows to verify the data structure
print("Grouped data sample:")
print(grouped_data.head())
```

```python
# Create a scatter plot showing the correlation between visibility and severity
plt.figure(figsize=(12, 20))

# Create scatter plot with visibility on x-axis and severity on y-axis
sns.scatterplot(
    x='Visibility(mi)',
    y='Severity',
    hue='State',  # Color points by location
    size='Year',     # Size points by year
    sizes=(20, 200), # Set the range of point sizes
    alpha=0.7,       # Add some transparency
    data=grouped_data
)

# Add a trend line to show the overall correlation
sns.regplot(
    x='Visibility(mi)',
    y='Severity',
    data=grouped_data,
    scatter=False,   # Don't add more scatter points
    color='red',     # Make the line red
    line_kws={"linestyle": "--"}  # Make it a dashed line
)

plt.title('Correlation Between Visibility and Accident Severity')
plt.xlabel('Average Visibility')
plt.ylabel('Average Accident Severity')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

#### A bar plot displayed showing that higher visibility scores are associated with more severe accidents.

#### The scatter displayed showing that higher wind speed is associated with more severe accidents

#### Let's check the correlation between visibility and accident severity

```python
# Extract relevant columns and handle potential data issues
# First, ensure numeric columns are properly formatted
accident_records['Visibility(mi)'] = pd.to_numeric(accident_records['Visibility(mi)'], errors='coerce')
accident_records['Severity'] = pd.to_numeric(accident_records['Severity'], errors='coerce')
accident_records['Weather_Condition'] = pd.to_numeric(accident_records['Weather_Condition'], errors='coerce')
```

```python
# Group by location, year to compute average visibility and severity
grouped_data = accident_records.groupby(['State', 'Year']).agg({
    'Visibility(mi)': 'mean',
    'Weather_Condition': 'mean',
    'Severity': 'mean'
}).reset_index()
```

```python
# Print the first few rows to verify the data structure
print("Grouped data sample:")
print(grouped_data.head())
```

```python
# Create a correlation matrix showing the relationship between weather clarity and accident severity
# Select only the relevant columns for correlation
correlation_data = grouped_data[['Visibility(mi)', 'Weather_Condition', 'Severity']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr(method='pearson')
print("Correlation Matrix:\n")
print(correlation_matrix)
```

```python
# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True,
    fmt='.2f',
    cmap='Blues',
    cbar=True,
    square=True,
    linewidths=0.5,
    linecolor='black'
)
plt.title('Correlation Between Weather Variables and Accident Severity')
plt.tight_layout()
plt.show()

# Extract specific correlation values
vis_sev_corr = correlation_matrix.loc['Visibility(mi)', 'Severity']
weather_sev_corr = correlation_matrix.loc['Weather_Condition', 'Severity']

print(f"\nCorrelation between Visibility and Severity: {vis_sev_corr:.3f}")
print(f"Correlation between Weather Clarity and Severity: {weather_sev_corr:.3f}")
```

#### The correlation matrix displayed showing that higher weather clarity (better visibility) is positively correlated with more severe accidents.


#### Let's Analyze whether worse weather leads to more severe crashes

```python
# Ensure numeric columns are properly formatted
accident_records['Visibility(mi)'] = pd.to_numeric(accident_records['Visibility(mi)'], errors='coerce')
accident_records['Severity'] = pd.to_numeric(accident_records['Severity'], errors='coerce')
```

```python
# Extract relevant columns
data_for_plot = accident_records[['State', 'Year', 'Visibility(mi)', 'Severity', 'Weather_Condition', 'Hour']]
```

```python
# Group by State and Year to compute average visibility
grouped_by_location = data_for_plot.groupby(['State', 'Year']).agg({'Visibility(mi)': 'mean'}).reset_index()
grouped_by_location.rename(columns={'Visibility(mi)': 'AverageVisibility'}, inplace=True)
```

```python
# Ensure 'Hour' column is included
if 'Hour' not in grouped_by_location.columns:
    grouped_by_location['Hour'] = accident_records.groupby(['State', 'Year'])['Hour'].mean().values

# Use correct DataFrame in line plot
plt.figure(figsize=(12, 8))
sns.lineplot(x='Hour', y='AverageVisibility', data=grouped_by_location)
plt.title('Weather Clarity by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Average Visibility Score (1: Good, 3: Poor)')
plt.xticks(range(24), [f"{h}:00" for h in range(24)])
plt.show()
```

#### The line chart displayed showing that higher visibility scores indicate better weather conditions and lower accident severity.
